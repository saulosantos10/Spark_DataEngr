Realizando Leitura de dados dos arquivos csv's disponibilizados
obs.: foi acrescentado o campo .option("sep",";") pois o arquivo original possui separação por ";"

```
# Read datas
dfJCS = spark.read \
        .option("inferSchema","True") \
        .option("header","True") \
        .option("sep",";") \
        .csv("dbfs:/FileStore/shared_uploads/jonathan.sales@aluno.faculdadeimpacta.com.br/jcs_2020.csv")
dfSCI = spark.read \
        .option("inferSchema","True") \
        .option("header","True") \
        .option("sep",";") \
        .csv("dbfs:/FileStore/shared_uploads/jonathan.sales@aluno.faculdadeimpacta.com.br/scimagojr_2020.csv")
```

Tratando o arquivo JCS

```
# Apresentando colunas e tipos de dados
display(dfJCS.dtypes)
```

```
# dropando colunas irrelevantes
newDfJCS = dfJCS.drop(*('_c3','_c6','_c7'))
# renomenado colunas df JCS
newDfJCSv2 = newDfJCS.withColumnRenamed("Full Journal Title","Title_JCS").withColumnRenamed("Total Cites","Total_Cites").withColumnRenamed("Journal Impact Factor","Journal_Impact_Factor").withColumnRenamed("Eigenfactor Score","Eigenfactor_Score").withColumnRenamed('Rank','Rank_JCS')
# substituindo virgula por ponto
from pyspark.sql.functions import *
newDfJCSv3 = newDfJCSv2.withColumn('Total_Cites', regexp_replace('Total_Cites',',','.'))
display(newDfJCSv3)
```

```
# Alterando tipo de dado de varias colunas
from pyspark.sql.types import StringType, DateType, FloatType
newDFJCSv4 = newDfJCSv3.withColumn("Total_Cites",newDfJCSv3['Total_Cites'].cast(FloatType())) \
             .withColumn('Journal_Impact_Factor',newDfJCSv3['Journal_Impact_Factor'].cast(FloatType())) \
             .withColumn('Eigenfactor_Score',newDfJCSv3['Eigenfactor_Score'].cast(FloatType()))
newDFJCSv4.dtypes
display(newDFJCSv4)
```

===========================================================================================================================================================================================

Tratando arquivo SCI Mago

```
# Apresentando colunas e seus tipos de dados
display(dfSCI.dtypes)
```

```
# Renomeando colunas SCI Mago
newDfSCI = dfSCI.withColumnRenamed("SJR Best Quartile","SJR_Best_Quartile").withColumnRenamed('H index','H_index').withColumnRenamed('Total Docs. (2020)','Total_Docs_(2020)').withColumnRenamed('Total Docs. (3years)','Total_Docs_3_years').withColumnRenamed('Total Refs.','Total_Refs').withColumnRenamed('Total Cites (3years)','Total_Cites_3_years').withColumnRenamed('Citable Docs. (3years)','Citable_Docs_3_years').withColumnRenamed('Cites / Doc. (2years)','Cites_Doc_2_years').withColumnRenamed('Ref. / Doc.','Ref_Doc').withColumnRenamed('Title','Title_SCI').withColumnRenamed('Rank','Rank_SCI')

# Substituindo virgula por ponto
from pyspark.sql.functions import *
newDfSCIv2 = newDfSCI.withColumn('SJR', regexp_replace('SJR',',','.')).withColumn('Cites_Doc_2_years', regexp_replace('Cites_Doc_2_years',',','.')).withColumn('Ref_Doc', regexp_replace('Ref_Doc',',','.'))
display(newDfSCIv2)


# substituindo tipos de dados
from pyspark.sql.types import StringType, DateType, FloatType
newDfSCIv3 = newDfSCIv2.withColumn("SJR",newDfSCIv2['SJR'].cast(FloatType())) \
             .withColumn('Cites_Doc_2_years',newDfSCIv2['Cites_Doc_2_years'].cast(FloatType())) \
             .withColumn('Ref_Doc',newDfSCIv2['Ref_Doc'].cast(FloatType()))
```

=========================================================================================================================================

Realizando junção das Tabelas

```
OutPutFinal = newDFJCSv4.join(newDfSCIv3, newDFJCSv4.Title_JCS == newDfSCIv3.Title_SCI,"leftouter" )
display(OutPutFinal)
```
